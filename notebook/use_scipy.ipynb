```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# 生成一些带噪声的模拟数据
np.random.seed(0)
x = np.linspace(0, 100, 100)
y = np.sin(x / 10.) + np.random.normal(0, 0.2, x.shape)

# 未处理的数据
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Original Data')

# 被高斯滤波处理过的数据|
sigma = 0.5
y_smeared = gaussian_filter1d(y, sigma=sigma)
plt.plot(x, y_smeared, label='Gaussian Smeared Data', linestyle='--')

plt.legend()
plt.show()
```

**gaussian_filter1d 用到的展宽值越大，涂抹后的原图越圆滑，可能越偏离原本的数据点**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# 生成一些带噪声的模拟数据
np.random.seed(0)
x = np.linspace(0, 100, 100)
y = np.sin(x / 10.) + np.random.normal(0, 0.2, x.shape)

# 设置要测试的 sigma 值
sigmas = [0.01, 0.5, 1.0, 2.0, 5.0, 10.0]

# 创建一个 2 行 3 列的子图布局
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()  # 将 axes 展平以便于迭代


# 循环绘制不同 sigma 值处理后的数据
for sigma, ax in zip(sigmas, axes[0:]):
    y_smeared = gaussian_filter1d(y, sigma=sigma)
    ax.plot(x, y, label='Original Data')
    ax.plot(x, y_smeared, label=f'Gaussian Smeared Data (sigma={sigma})', linestyle='--')
    ax.set_title(f'Gaussian Smeared Data (sigma={sigma})')

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```
