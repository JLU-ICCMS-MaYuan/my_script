### 229号空间群在所有结构中的比例

```python
total_count = 100
specified_paras = ["para1", "para2", "para3", "para4", "para5", "para6", "para7", "para8", "para9", "calypso1"]

poscar_229_count = [42, 12, 4, 0, 4, 2, 0, 0, 0]
poscar_229_ratio = [round(c/total_count*100, 1) for c in poscar_229_count]
poscar_rest_count = [total_count - c for c in poscar_229_count]

contcar_229_count = [42, 12, 5, 0, 4, 3, 0, 0, 0]
contcar_229_ratio = [round(c/total_count*100, 1) for c in contcar_229_count]
contcar_rest_count = [total_count - c for c in contcar_229_count]


calypso_poscar_229_count = [2]
calypso_poscar_229_ratio = [round(c/total_count*100, 1) for c in calypso_poscar_229_count]
calypso_poscar_rest_count = [total_count - c for c in calypso_poscar_229_count]

calypso_contcar_229_count = [2]
calypso_contcar_229_ratio = [round(c/total_count*100, 1) for c in calypso_contcar_229_count]
calypso_contcar_rest_count = [total_count - c for c in calypso_contcar_229_count]
print(calypso_contcar_229_count)
```

```python
fig, axs = plt.subplots(1 , 1, figsize=(10,4))

# 我的方法
# POSCAR 的对称性统计
for position, _ in enumerate(poscar_229_count):
    # 如果想根据每一个柱状图的情况设置每一个柱体的标签，就在axs[i].bar这里设置label
    axs.bar(position, poscar_229_count[position], width=0.3, color="blue")
    axs.text(position, poscar_229_count[position]*1.05, str(poscar_229_count[position]), horizontalalignment='center', weight='bold', fontsize=10,)
    # 堆积的要点在于绘制时候的bottom关键字，这决定了柱状图的绘制起点，因此第二段的bottom是第一段的数值，而第三段则为第一、二段相加
    axs.bar(position, poscar_rest_count[position], width=0.3, color="cornflowerblue", bottom=poscar_229_count[position])

# CONTCAR 的对称性统计
for position, _ in enumerate(contcar_229_count):
    axs.bar(position+0.3, contcar_229_count[position], width=0.3, color="green")
    axs.text(position+0.3, contcar_229_count[position]*1.05, str(contcar_229_count[position]), horizontalalignment='center', weight='bold', fontsize=10,)
    axs.bar(position+0.3, contcar_rest_count[position], width=0.3, color="lightgreen", bottom=contcar_229_count[position])

# CALYPSO 的方法
# POSCAR 的对称性统计
for position, _ in enumerate(calypso_poscar_229_count):
    axs.bar(position+len(poscar_229_count), calypso_poscar_229_count[position], width=0.3, color="blue")
    axs.text(position+len(poscar_229_count), calypso_poscar_229_count[position]*1.05, str(calypso_poscar_229_count[position]), horizontalalignment='center', weight='bold', fontsize=10,)
    axs.bar(position+len(poscar_229_count), calypso_contcar_rest_count[position], width=0.3, color="cornflowerblue", bottom=calypso_poscar_229_count[position])

# CONTCAR 的对称性统计
print(calypso_contcar_229_count)
for position, _ in enumerate(calypso_contcar_229_count):
    axs.bar(position+len(poscar_229_count)+0.3, calypso_contcar_229_count[position], width=0.3, color="green")
    axs.text(position+len(poscar_229_count)+0.3, calypso_contcar_229_count[position]*1.05, str(calypso_contcar_229_count[position]), horizontalalignment='center', weight='bold', fontsize=10,)
    axs.bar(position+len(poscar_229_count)+0.3, calypso_contcar_rest_count[position], width=0.3, color="lightgreen", bottom=calypso_contcar_229_count[position])


axs.set_xticks([specified_paras.index(i)+0.15 for i in specified_paras])
axs.set_xticklabels(specified_paras)

# 如果只想设置一个label，那就再lenged这里设置包含一个元素的列表即可
axs.legend(["The Number of crystal with spg-229"], loc='upper left',  shadow=True)
axs.set_ylim(0, total_count+total_count*0.40)
axs.set_title("Efficiency Comparison of CALYPSO and My Method")
```
