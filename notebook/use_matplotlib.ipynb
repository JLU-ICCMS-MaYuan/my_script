```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
specified_spgnum = 229
specified_count = [8, 10, 12]
specified_paras = ["para1", "para2", "para3"]
specified_ratio = [round(c/total_count*100, 1) for c in specified_count]
total_count = 500
rest_count = [total_count - c for c in specified_count]
```

```python
fig, axs = plt.subplots(1 , 1, figsize=(10,4))


for id, _ in enumerate(specified_count):
    # 如果想根据每一个柱状图的情况设置每一个柱体的标签，就在axs[i].bar这里设置label
    axs.bar(id, specified_count[id], width=0.3, color="orangered")
    axs.text(id, specified_count[id]+specified_count[id]*0.20, str(specified_count[id]), horizontalalignment='center', weight='bold', fontsize=10,)
    # 堆积的要点在于绘制时候的bottom关键字，这决定了柱状图的绘制起点，因此第二段的bottom是第一段的数值，而第三段则为第一、二段相加
    axs.bar(id, rest_count[id], width=0.3, color="dodgerblue", bottom=specified_count[id])
   

axs.set_xticks([specified_paras.index(i) for i in specified_paras])
axs.set_xticklabels(specified_paras)

# 如果只想设置一个label，那就再lenged这里设置包含一个元素的列表即可
axs.legend(["The Number of CaH6"], loc='upper left',  shadow=True)
axs.set_ylim(0, total_count+total_count*0.40)
axs.set_title("My Method")
```
