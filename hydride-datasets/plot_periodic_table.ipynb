```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_squares(ax):
    # 设置正方形的左上角坐标
    # No3period = [np.array([0.0, 0.2]) + np.array([0.1+sqr, 0]) for sqr in np.arange(0, 1.8, 0.1)]
    No2period = [[sqr, 0.79] for sqr in np.arange(0, 1.98, 0.11)]
    No3period = [[sqr, 0.68] for sqr in np.arange(0, 1.98, 0.11)]
    for x, y in No2period:
        ax.add_patch(plt.Rectangle((x, y), 0.1, 0.1, edgecolor='black', fill=False))
    for x, y in No3period:
        ax.add_patch(plt.Rectangle((x, y), 0.1, 0.1, edgecolor='black', fill=False))
    # ax.set_xlim(0, 1)
    # ax.set_ylim(0, 1)

fig, ax = plt.subplots()
draw_squares(ax)
plt.axis('off') # 关闭坐标轴
plt.show()
```

```python
def draw_squares(ax, column, infos, valEle):
    # 设置正方形的左上角坐标
    for coords, info, val in zip(column, infos, valEle):
        ax.add_patch(plt.Rectangle((coords[0], coords[1]), 0.1, 0.1, edgecolor='black', fill=False))
        ax.text(coords[0]+0.05, coords[1]+0.06, info, ha='center', va='center', fontsize=12)
        ax.text(coords[0]+0.05, coords[1]+0.02, val,  ha='center', va='center', fontsize=7.5)
    
    # 函数设置了 x 轴和 y 轴的范围
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.tick_params(axis='both', which='major', labelsize=0.5)
    ax.tick_params(axis='both', which='minor', labelsize=0.1)



fig, ax = plt.subplots(figsize=(8, 8), dpi=300)
ax.grid(linestyle='dotted', color='gray')

No1A  = [[0.10, sqr] for sqr in np.arange(0.77, 0.10, -0.11)]
infos = ['H', 'Li', 'Na', 'K', 'Rb', 'Cs', 'Fr']
valEle= ['$1s^1$','$2s^1$', '$3s^1$', '$4s^1$', '$5s^1$', '$6s^1$', '$7s^1$']
draw_squares(ax, No1A, infos, valEle)

No2A  = [[0.21, sqr] for sqr in np.arange(0.66, 0.10, -0.11)]
infos = ['Be', 'Mg', 'Ca', 'Sr', 'Ba', 'Ra']
valEle= ['$2s^2$','$3s^2$', '$4s^2$', '$5s^2$', '$6s^2$', '$7s^2$']
draw_squares(ax, No2A, infos, valEle)

No3B  = [[0.32, sqr] for sqr in np.arange(0.44, 0.10, -0.11)]
infos = ['Sc', 'Y', 'La', 'Ac']
valEle= ['$3p^6 3d^1$ \n $4s^2$', '$3d^{10} 4p^6$ \n $5s^2 4d^1$', '$4d^{10} 5p^6 $\n $6s^2 5d^1$', '$4f^{14} 5d^{10}$ \n $6p^6 7s^2 6d^1$']
draw_squares(ax, No3B, infos, valEle)

No3B  = [[0.43, sqr] for sqr in np.arange(0.44, 0.10, -0.11)]
infos = ['Ti', 'Zr', 'Hf']
valEle= ['$3p^6 4s^2$ \n $3d^1$', '$3d^{10} 4p^6$ \n $5s^2 $', '$4d^{10} 5p^6 $\n $5d^1 6s^2$', '$4f^{14} 5d^{10}$ \n $6p^6 7s^2 6d^1$']
draw_squares(ax, No3B, infos, valEle)
# plt.axis('off') # 关闭坐标轴
plt.show()
```
