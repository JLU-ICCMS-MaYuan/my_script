```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_squares(ax, column: list[str], infos: list[str], valEle: list[str], atomicNumber: list[int]):
    """
    
    
        ax: matplotlibax, 
        column: ()
        infos:  
        valEle: 
    """
    # 
    for coords, info, val, atomN in zip(column, infos, valEle, atomicNumber):
        if info in ['Y', 'La', 'Li', 'Sc', 'Ce', 'Ca', 'Mg']:
            ax.add_patch(plt.Rectangle((coords[0], coords[1]), 0.1, 0.1, edgecolor='black', fill=True, facecolor='red'))
            ax.text(coords[0]+0.05, coords[1]+0.06, info, ha='center', va='center', fontsize=15)
            ax.text(coords[0]+0.05, coords[1]+0.02, val,  ha='center', va='center', fontsize=8)
            ax.text(coords[0]+0.02, coords[1]+0.085, atomN, ha='center', va='center', fontsize=7)
        else:   
            ax.add_patch(plt.Rectangle((coords[0], coords[1]), 0.1, 0.1, edgecolor='black', fill=False))
            ax.text(coords[0]+0.05, coords[1]+0.06, info, ha='center', va='center', fontsize=15)
            ax.text(coords[0]+0.05, coords[1]+0.02, val,  ha='center', va='center', fontsize=8)
            ax.text(coords[0]+0.02, coords[1]+0.085, atomN, ha='center', va='center', fontsize=7)
    
    # 
    ax.xaxis.set_tick_params(width=2, size=5, labelsize=10)
    ax.yaxis.set_tick_params(width=2, size=5, labelsize=10)

def LaSys_AcSys(ax, column: list[str], infos: list[str], valEle: list[str], atomicNumber: list[int]):
    # LaAc
    for coords, info, val, atomN in zip(column, infos, valEle, atomicNumber):
        ax.add_patch(plt.Rectangle((coords[0], coords[1]), 0.1, 0.1, edgecolor='black', fill=False))
        ax.text(coords[0]+0.05, coords[1]+0.06, info, ha='center', va='center', fontsize=15)
        ax.text(coords[0]+0.05, coords[1]+0.02, val,  ha='center', va='center', fontsize=8)
        ax.text(coords[0]+0.02, coords[1]+0.085, atomN, ha='center', va='center', fontsize=7)

fig, ax = plt.subplots(figsize=(18, 11.2), dpi=300)
#  x  y 
ax.set_xlim(0, 2.25)
ax.set_ylim(0, 1.4)
ax.grid(linestyle='dotted', color='gray')

No1A  = [[0.10, sqr] for sqr in np.arange(1.1, 0.43, -0.11)]
infos = ['H', 'Li', 'Na', 'K', 'Rb', 'Cs', 'Fr']
atomN = [1, 3, 11, 19, 37, 55, 87]
valEle= ['$1s^1$','$2s^1$', '$3s^1$', '$4s^1$', '$5s^1$', '$6s^1$', '$7s^1$']
draw_squares(ax, No1A, infos, valEle, atomN)

No2A  = [[0.21, sqr] for sqr in np.arange(0.99, 0.43, -0.11)]
infos = ['Be', 'Mg', 'Ca', 'Sr', 'Ba', 'Ra']
atomN = [4, 12, 20, 38, 56, 88]
valEle= ['$2s^2$','$3s^2$', '$4s^2$', '$5s^2$', '$6s^2$', '$7s^2$']
draw_squares(ax, No2A, infos, valEle, atomN)

LaCol = [[0.32, 0.55], [0.32, 0.44]]
infos = ['La', 'Ac']
atomN = [57, 89]
valEle= ['$6s^2 5d^1$', '$7s^2 6d^1$']
draw_squares(ax, LaCol, infos, valEle, atomN)

Col   = [[0.43, 0.55], [0.43, 0.44]]
infos = ['...', '...']
atomN = ['...', '...']
valEle= ['...', '...']
draw_squares(ax, Col, infos, valEle, atomN)

LaSys  = [[sqr, 0.27] for sqr in np.arange(0.32, 1.97, 0.11)]
LaInfos=['La', 'Ce', 'Pr', 'Nd', 'Pm', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Tm', 'Yb', 'Lu']
atomN  = [i for i in range(57, 71)]
LaValEle =['$6s^2 5d^1$', '$6s^2 4f^1 5d^1$', '$6s^2 4f^3$', '$6s^2 4f^4$', '$6s^2 4f^5$', '$6s^2 4f^6$', '$6s^2 4f^7$', '$6s^2 4f^7 5d^1$', '$6s^2 4f^9$', '$6s^2 4f^{10}$', '$6s^2 4f^{11}$', '$6s^2 4f^{12}$', '$6s^2 4f^{13}$', '$6s^2 4f^{14}$', '$6s^2 4f^{14} 5d^1$']
LaSys_AcSys(ax, LaSys, LaInfos, LaValEle, atomN)

AcSys  = [[sqr, 0.16] for sqr in np.arange(0.32, 1.97, 0.11)]
AcInfos=['Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr']
atomN  = [i for i in range(89, 103)]
AcValEle =['$7s^2 6d^1$', '$7s^2 6d^2$', '$7s^2 5f^2 6d^1$', '$7s^2 5f^3 6d^1$', '$7s^2 5f^4 6d^1$', '$7s^2 5f^6$', '$7s^2 5f^7$', '$7s^2 5f^7 6d^1$', '$7s^2 5f^9$', '$7s^2 5f^{10}$', '$7s^2 5f^{11}$', '$7s^2 5f^{12}$', '$7s^2 5f^{13}$', '$7s^2 5f^{14}$', '$7s^2 5f^{14} 7p^1$']
LaSys_AcSys(ax, AcSys, AcInfos, AcValEle, atomN)

No3B  = [[0.54, sqr] for sqr in np.arange(0.77, 0.43, -0.11)]
infos = ['Sc', 'Y', 'Lu', 'Lr']
atomN = [21, 39, 71, 103]
valEle= ['$4s^2 3d^1$', '$5s^2 4d^1$', '$6s^2 4f^{14} 5d^1$', '$7s^2 5f^{14} 7p^1$']
draw_squares(ax, No3B, infos, valEle, atomN)

No4B  = [[0.65, sqr] for sqr in np.arange(0.77, 0.43, -0.11)]
infos = ['Ti', 'Zr', 'Hf', 'Rf']
atomN = [22, 40, 72, 104]
valEle= ['$4s^2 3d^2$', '$5s^2 4d^2$', '$6s^2 4f^{14} 5d^2$', '$7s^2 5f^{14} 6d^3$']
draw_squares(ax, No4B, infos, valEle, atomN)

No5B  = [[0.76, sqr] for sqr in np.arange(0.77, 0.43, -0.11)]
infos = ['V', 'Nb', 'Ta', 'Db']
atomN = [23, 41, 73, 105]
valEle= ['$4s^3 3d^3$', '$5s^1 4d^4$', '$6s^2 4f^{14} 5d^3$', '$7s^2 5f^{14} 6d^2$']
draw_squares(ax, No5B, infos, valEle, atomN)

plt.axis('off') # 
plt.show()
```
